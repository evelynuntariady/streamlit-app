```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
```

## Read Data

```python
data = pd.read_csv('Iris.csv')
```

```python
cols = data.columns
```

## Encoding

```python
data['Species'].replace('Iris-setosa', 0, inplace=True)
data['Species'].replace('Iris-versicolor', 1, inplace=True)
data['Species'].replace('Iris-virginica', 2, inplace=True)
```

## Split Data

```python
x_temp = data.drop('Species', axis = 1)
x = x_temp.drop('Id', axis = 1)
y = data['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 42)
```

```python
x_train.shape
```

```python
y_train.shape
```

## Modeling

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(random_state = 42)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
